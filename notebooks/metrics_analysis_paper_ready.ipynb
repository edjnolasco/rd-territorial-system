Notebook paper-ready para analizar métricas del sistema.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../reports/metrics/metrics.csv')
df.head()
```

```python
df.info()
```

## Resolver outcome distribution

```python
resolver_df = df[df['endpoint'].str.contains('/resolve', na=False)]
resolver_df['result_type'].value_counts()
```

```python
resolver_df['result_type'].value_counts().plot(kind='bar')
plt.title('Resolver Outcome Distribution')
plt.show()
```

## Latency analysis

```python
df['latency_ms'].plot(kind='hist', bins=20)
plt.title('Latency Distribution')
plt.show()
```

## Latency by endpoint

```python
df.groupby('endpoint')['latency_ms'].mean().plot(kind='bar')
plt.title('Latency by Endpoint')
plt.show()
```

## Client usage

```python
df['client_id'].value_counts().plot(kind='bar')
plt.title('Requests per Client')
plt.show()
```

## Paper summary

```python
summary = {
    'total_requests': len(df),
    'resolver_requests': len(resolver_df),
    'matched_pct': (resolver_df['result_type'] == 'matched').mean() * 100,
    'ambiguous_pct': (resolver_df['result_type'] == 'ambiguous').mean() * 100,
    'no_match_pct': (resolver_df['result_type'] == 'no_match').mean() * 100,
    'mean_latency': df['latency_ms'].mean(),
    'median_latency': df['latency_ms'].median()
}
summary
```
